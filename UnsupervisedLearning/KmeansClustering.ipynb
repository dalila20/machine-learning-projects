```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
mean_1 = [np.random.randint(100), np.random.randint(100)]
mean_2 = [np.random.randint(100), np.random.randint(100)]

# Diagonal covariance
cov = [[100,0], [0,100]]

x1, y1 = np.random.multivariate_normal(mean_1, cov, 100).T
x2, y2 = np.random.multivariate_normal(mean_2, cov, 100).T

x = np.append(x1, x2)
y = np.append(y1, y2)

plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
X = np.array(list(zip(x, y)))
```

```python
# Model
kmeans = KMeans(n_clusters=2)
```

```python
# Fitting the model
kmeans.fit(X)
```

```python
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print(labels)
print(centroids)
```

```python
colors = ['g.', 'r.']

for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize=10)

plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=150, zorder=10)
plt.show()
```
